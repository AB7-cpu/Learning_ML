**Capping**

```python
#Capping

upper_limit = df["cgpa"].mean()+3*df["cgpa"].std()
lower_limit = df["cgpa"].mean()-3*df["cgpa"].std()

print(upper_limit)
print(lower_limit)

df["cgpa"] = np.where(
    df["cgpa"]>upper_limit,
    upper_limit,
    np.where(
        df["cgpa"]<lower_limit,
        lower_limit,
        df["cgpa"]
    )
)
```
