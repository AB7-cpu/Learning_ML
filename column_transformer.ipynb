```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset//covid_toy.csv')
```

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:5], df.iloc[:,-1], test_size=0.2, random_state=0)
X_train.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder

ct = ColumnTransformer(transformers=[
    ('tf1',SimpleImputer(), ['fever']),
    ('tf2',OneHotEncoder(sparse_output=False, dtype=np.int32, drop='first'), ['gender', 'city']),
    ('tf3',OrdinalEncoder(categories=[['Mild', 'Strong']]), ['cough'])
], remainder='passthrough')

ct.fit_transform(X_train)
```
