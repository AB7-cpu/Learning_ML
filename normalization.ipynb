```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('dataset//wine_data.csv', header=None, usecols=[0,1,2])
df.columns = ['Class label', 'Alcohol', 'Malic Acid']
```

```python
df.head()
```

```python
color_dict = {'Alcohol': 'red', 'Malic_Acid':'green', 'class_label':'blue'}
sns.scatterplot(df[['Alcohol', 'Malic Acid', 'Class label']])#, df['Malic Acid']# hue= df['Class label'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class label', axis = 1), df['Class label'], test_size=0.3, random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
np.round(X_train_scaled.describe(), 1)
```
